```python
from selenium import webdriver
from selenium .webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.select import Select
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

# using selenium to automate the website

```python
s=Service(r'C:\Users\ayush\Desktop\AYUSH FILES\crome driver\chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get("https://www.imdb.com/")

#dropdown
time.sleep(1)
dropdown=driver.find_element(by=By.CLASS_NAME,value="ipc-icon--arrow-drop-down")
dropdown.click()
time.sleep(1)

#advanced search from drop down menu
element=driver.find_element(by=By.LINK_TEXT,value="Advanced Search")
element.click()
time.sleep(1)

#advance title search
element2=driver.find_element(by=By.LINK_TEXT,value="Advanced Title Search")
element2.click()
time.sleep(1)

#feature film
feature_film=driver.find_element(by=By.ID,value="title_type-1")
feature_film.click()
time.sleep(1)

#tv_movie
tv_movie=driver.find_element(by=By.ID,value="title_type-2")
tv_movie.click()
time.sleep(1)

#min date
min_date=driver.find_element(by=By.NAME,value="release_date-min")
min_date.click()
min_date.send_keys("1990")


#max_date
max_date=driver.find_element(by=By.NAME,value="release_date-max")
max_date.click()
max_date.send_keys("2021")


#rating_min
rating_min=driver.find_element(by=By.NAME,value="user_rating-min")
rating_min.click()
select_min=Select(rating_min)
select_min.select_by_visible_text("1.0")
time.sleep(1)

#rating_max
rating_max=driver.find_element(by=By.NAME,value="user_rating-max")
rating_max.click()
select_max=Select(rating_max)
select_max.select_by_visible_text("10")


#Oscar_Nominated
oscar_Nominated=driver.find_element(by=By.ID,value="groups-7")
oscar_Nominated.click()

#select language
select_language=driver.find_element(by=By.NAME,value="languages")
select_english=Select(select_language)
select_english.select_by_visible_text("English")


#select display
select_display=driver.find_element(by=By.ID,value="search-count")
select_100= Select(select_display)
select_100.select_by_index(1)

#click search
click_search=driver.find_element(by=By.XPATH,value="/html/body/div[2]/div/div[2]/div[3]/form/div/p[3]/button")
click_search.click()


#current
url=driver.current_url

############################################################################




#get requests
r=requests.get(url)

#status code
r.status_code


#soup object
htmlcontent=r.content
soup=bs(htmlcontent,"html.parser")
#print(soup.prettify())


#to find:-
#title
#year
#duration
#genre
#rating


#result items(starting point)
list_items=soup.find_all("div",{'class':"lister-item"})
len(list_items)



#title
#list_items[0].find('h3').find('a').get_text()
#year
#list_items[0].find('h3').find('span',{'class':'lister-item-year'}).get_text().replace('(','').replace(')','')
#duration
#list_items[0].find('span',{'class':'runtime'}).get_text()
#genre
#list_items[0].find('span',{'class':'genre'}).get_text().strip()
#rating
#list_items[0].find('div',{'class':'inline-block ratings-imdb-rating'}).get_text().strip()




#using list comprehension
movie_title=[result.find('h3').find('a').get_text() for result in list_items]
movie_year=[result.find('h3').find('span',{'class':'lister-item-year'}).get_text().replace('(','').replace(')','') for result in list_items]
movie_duration=[result.find('span',{'class':'runtime'}).get_text() for result in list_items]
movie_genre=[result.find('span',{'class':'genre'}).get_text().strip() for result in list_items]
movie_rating=[result.find('div',{'class':'inline-block ratings-imdb-rating'}).get_text().strip() for result in list_items]


#create dataframe
imdb_df=pd.DataFrame({'Movie Title':movie_title, 'Year':movie_year,'Duration':movie_duration,'Genre':movie_genre,'Rating':movie_rating})
imdb_df


#output in excel
imdb_df.to_excel('imd_project.xlsx',index=False)




```
